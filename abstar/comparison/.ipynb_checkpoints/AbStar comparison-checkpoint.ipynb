```python
import json

import numpy as np
import pandas as pd
```

##Parse AbStar's JSON output

```python
abstar_all_ds = {}
abstar_top_d = {}
with open('../test_data/test1k_output/test_HDx_1k.json') as f:
    for line in f:
        j = json.loads(line.rstrip('\n'))
        if 'd_gene' not in j:
            continue
        seq_id = j['seq_id']
        top_d = (j['d_gene']['gene'], j['d_gene']['score'])
        all_ds = [(d['gene'], d['score']) for d in j['d_gene']['others']]
        abstar_top_d[seq_id] = top_d
        abstar_all_ds[seq_id] = all_ds
```

##Parse IMGT and IgBLAST output

```python
igblast_top_d = {}
with open('../comparison/HDx1k_igblast_output.txt') as f:
    for line in f:
        s = line.split('\t')
        seq_id = s[0]
        d_gene = s[2].split('*')[0]
        igblast_top_d[seq_id] = d_gene

imgt_top_d = {}
with open('../comparison/HDx1k_imgt_output.txt') as f:
    for line in f:
        s = line.split('\t')
        seq_id = s[0]
        d_gene = s[2].split('*')[0].split(' ')[1] if s[2] != '' else ''
        imgt_top_d[seq_id] = d_gene
```

```python
d_genes = {'abstar_top': abstar_top_d, 
           'abstar_all': abstar_all_ds, 
           'igblast': igblast_top_d, 
           'imgt': imgt_top_d}

df = pd.DataFrame(d_genes)
df.fillna('none', inplace=True)
print df.to_string()
```

```python
imgt_top_match = 0
imgt_any_match = 0
imgt_no_match = 0



for row in df.iterrows():
    all_abstar = row[1][0]
    top_abstar = row[1][1]
    top_igblast = row[1][2]
    top_imgt = row[1][3]
    
    if top_imgt == top_abstar[0]:
        imgt_top_match += 1
    elif top_imgt in [d[0] for d in all_abstar]:
        imgt_any_match += 1
    else:
        if top_imgt == '' or top_abstar[0] == 'none' or top_abstar[0] == 'n' or top_abstar[0] == None:
            continue
        imgt_no_match += 1
imgt_total = sum([imgt_top_match, imgt_any_match, imgt_no_match])
print 'IMGT top match\t{}'.format(100. * imgt_top_match / imgt_total)
print 'IMGT any match\t{}'.format(100. * imgt_any_match / imgt_total)
print 'IMGT no match\t{}'.format(100. * imgt_no_match / imgt_total)
```

```python
igblast_top_match = 0
igblast_any_match = 0
igblast_no_match = 0



for row in df.iterrows():
    all_abstar = row[1][0]
    top_abstar = row[1][1]
    top_igblast = row[1][2]
    top_imgt = row[1][3]
    
    if top_igblast == top_abstar[0]:
        igblast_top_match += 1
    elif top_igblast in [d[0] for d in all_abstar]:
        igblast_any_match += 1
    else:
        if top_igblast == '' or top_abstar[0] == 'none' or top_abstar[0] == 'n' or top_abstar[0] == None:
            continue
        igblast_no_match += 1
igblast_total = sum([igblast_top_match, igblast_any_match, igblast_no_match])
print 'IgBLAST top match\t{}'.format(100. * igblast_top_match / igblast_total)
print 'IgBLAST any match\t{}'.format(100. * igblast_any_match / igblast_total)
print 'IgBLAST no match\t{}'.format(100. * igblast_no_match / igblast_total)
```

```python
imgt_top_match_scores = []
imgt_any_match_scores = []
imgt_no_match_scores = []



for row in df.iterrows():
    all_abstar = row[1][0]
    top_abstar = row[1][1]
    top_igblast = row[1][2]
    top_imgt = row[1][3]
    
    if top_imgt == top_abstar[0]:
        best_non_allele = [a[1] for a in all_abstar if a[0] != top_abstar[0]]
        imgt_top_match_scores.append(top_abstar[1] - max(best_non_allele))
        #imgt_top_match_scores.append(100. * (top_abstar[1] - all_abstar[0][1]) / top_abstar[1])
    elif top_imgt in [d[0] for d in all_abstar]:
        match_score = [a[1] for a in all_abstar if top_imgt == a[0]][0]
        imgt_any_match_scores.append(top_abstar[1] - match_score)
        #imgt_any_match_scores.append(100. * (top_abstar[1] - match_score) / top_abstar[1])
    else:
        if all_abstar == 'none':
            continue
        if top_imgt == '' or top_abstar[0] == 'none' or top_abstar[0] == 'n' or top_abstar[0] == None:
            continue
        match_score = [a[1] for a in all_abstar][-1]
        imgt_no_match_scores.append(top_abstar[1] - match_score)
        #imgt_no_match_scores.append(100. * (top_abstar[1] - match_score) / top_abstar[1])
print('\n'.join([str(s) for s in imgt_top_match_scores]))
```
